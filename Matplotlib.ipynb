```python
from matplotlib import pyplot as plt
x=[2,3,6]
y=[2,4,5]
plt.plot(x,y)
plt.xlabel("x axis") #labelling
plt.ylabel("y axis")
plt.title("simple line chart")
plt.show()
```

```python
#Markers(to represent points in graph )
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,10,2)
y=x*2
plt.plot(x,y,marker='s',mec="r",mfc="b",ms="10") # (we can use d, p, _, |, + for different style of markers)
                                                 #(mec,mfc,ms are edge color,face color,size of markers)
plt.show()
```

```python
#lines (to give line style)
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,10,2)
y=x*2
x1=[5,1,15,2]
y1=[20,3,4,50]
plt.plot(x,y,x1,y1,ls="dotted",lw=5,c="k") #lw=linewidth,c=color,ls=line style. none is used for hiding line
plt.show()
```

```python
#font properties
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,10,2)
y=x*2
plt.plot(x,y)
f1={'family':'arial','size':20,'color':'red'}
f2={'family':'arial','size':15,'color':'blue'}
plt.grid(axis='both',ls='dotted',color='r',lw=2)  #to create grid layout,we can create grid on x,y axis also sepaartely
# we can use line style,width also for grid lines
plt.xlabel(" x values",fontdict=f1)
plt.ylabel(" y values",fontdict=f2)
plt.show()
```

```python
#subplots (to create  multiple plots)
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,10,2)
y=x*2
y1=[20,30,40,50,60]
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("plot1")

plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title("plot2")

plt.subplot(2,2,3)
plt.plot(x,y1)
plt.title("plot3")

plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("plot4")


plt.show()
```

```python
#legend 
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,10,2)
y=x*2
y1=[2,5,7,9,11]
plt.plot(x,y,label="squares")
plt.plot(x,y1,label="cubes")
plt.title("plot1")
#plt.legend() #without parameters
# with parameters
plt.legend(loc="upper right",facecolor="blue",edgecolor="yellow",framealpha=0.2,shadow=True)
plt.show()
```

```python
#scatter (to plot points without lines)
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,10,2)
y=x*2
y1=x*3
plt.scatter(x,y,color="black")
plt.scatter(x,y1,color="blue",marker='*')
plt.show()
```

```python
#colormap
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,10,2)
y=x*2
y1=x*3
sizes=[10,20,30,40,50]
colors=[2,4,6,7,8]
plt.scatter(x,y,s=sizes,c=colors,cmap="rainbow")
plt.scatter(x,y1,s=sizes,c=colors,cmap="rainbow",alpha=0.4)
plt.colorbar()
plt.show()
```

```python
#barplot
from matplotlib import pyplot as plt
import numpy as np
years=["2","3","4","5"]
cse=[200,300,400,450]
ece=[200,100,300,207]
it=[300,200,756,124]
w=0.2

#locations to plot
cse_bar=np.arange(len(years))
ece_bar=[i+w for i in cse_bar]
it_bar=[i+w for i in ece_bar]

plt.bar(cse_bar,cse,width=w,label="cse")
plt.bar(ece_bar,cse,width=w,label="ece")
plt.bar(it_bar,cse,width=w,label="it")
plt.legend()

plt.show()
```

```python
#pie chart
from matplotlib import pyplot as plt
s=["good","avg","poor","fail"]
v=[12,13,19,15]
plt.pie(v,labels=s,explode=[0,0.1,0,0.2],shadow=True,startangle=180) #explode is used to cut a part of it
plt.legend()
plt.show()
```

```python
#histogram
from matplotlib import pyplot as plt
marks=[10,30,50,80,77,88,90]
grade=[35,50,70,100]
plt.hist(marks,grade,histtype="stepfilled",rwidth=0.2,facecolor="k")
plt.xlim([10,100])
plt.xticks([0,35,70,100])
plt.show()
```
