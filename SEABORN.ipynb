```python
### seaborn is a python data visualization library based on matplotlib .it provides a high level interface for drawing attractive and informative statistical graphics ##
```

```python
## scatter plot ##
```

```python
import pandas as pd
sales=pd.read_excel("sales.xlsx")
sales.dropna()
sales.head()
```

```python
import seaborn as sns
## relplot : it finds relationship b/w two variables.##
sns.relplot('Sales','Profit',data=sales) 
```

```python
## hue: it takes column name for colour encoding##
sns.relplot(x='Sales',y='Profit',data=sales,hue='Order Priority') 
```

```python
sns.relplot(x='Sales',y='Profit',data=sales,hue='Order Priority',style='Order Priority')
```

```python
 sns.relplot(x='Sales',y='Profit',data=sales,size='Discount',sizes=(20,200))
```

```python
### LINE CHART ###
```

```python
sns.relplot(x='Sales',y='Profit',data=sales,kind='line')
```

```python
sns.relplot(x='Sales',y='Profit',data=sales,kind='line',hue='Ship Mode')
```

```python
## multiple charts ##  
sns.relplot(x="Sales",y="Profit",data=sales,col="Customer Segment")
```

```python
## col wrap is used for no of plots for each col ##
sns.relplot(x="Sales",y="Profit",data=sales,col="Customer Segment",col_wrap=2)
```

```python
## CATEGORICAL PLOT ## 
sns.catplot(x="Ship Mode",y="Sales",data=sales)
```

```python
## jitter= false ,it removes magnitude and jittering .it shows patterns clearly##
sns.catplot(x="Ship Mode",y="Sales",data=sales,jitter=False)
```

```python
### BOXPLOT:It tells about how our data is distributed and what the median is or outliers ##
sns.boxplot(x="Product Category",y="Shipping Cost",data=sales)
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",data=sales,kind='box')
```

```python
sns.boxplot(x="Product Category",y="Shipping Cost",hue="Ship Mode",data=sales)
```

```python
## for each row or col  it shows seperate graph if we use row or col ##
sns.catplot(x="Product Category",y="Shipping Cost",col="Ship Mode",data=sales,kind="box")
```

```python
## violin plot: It is like box plot and violint plot allows to visualize the distribution of a numeric variable for one or several groups. ##
sns.catplot(x="Product Category",y="Shipping Cost",data=sales,kind="violin")
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",hue="Ship Mode",data=sales,kind="violin")
```

```python
sns.barplot(x="Product Category",y="Shipping Cost",hue="Ship Mode",data=sales)
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",data=sales,kind="point")
```

```python
## distribution plots ##
sns.distplot(sales.Sales)
```

```python
sns.distplot(sales.Profit)
```

```python
## KDE(kernel density estimate) represents the data using a continuous probability density curve in one or more dimensions.
sns.distplot(sales.Profit,kde=False)
```

```python
## rug shows where most of the data points lies ##
sns.distplot(sales.Profit,hist=False,rug=True)
```

```python
sns.distplot(sales.Profit,kde=False,rug=True,bins=100)
```

```python
## kde plot ##
sns.kdeplot(sales.Profit,bw=1000) #bw =bandwidth
```
