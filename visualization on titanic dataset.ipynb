```python
import seaborn as sn
import pandas as pd
```

```python
sn.get_dataset_names()
```

```python
titanic=sn.load_dataset('titanic')
titanic.head()
```

```python
sn.distplot(titanic['fare'],color='blue',bins=10)
```

```python
sn.jointplot(x='survived',y='age',data=titanic,hue='sex')
```

```python
sn.jointplot(x='survived',y='age',data=titanic,kind='reg')
```

```python
sn.pairplot(titanic,hue='sex')
```

```python
sn.barplot(x='survived',y='age',data=titanic)
```

```python
sn.boxplot(x='sex',y='age',data=titanic,hue='survived')
```

```python
sn.violinplot(x='sex',y='age',data=titanic)
```

```python
sn.catplot(x='class',y='survived',data=titanic,kind='box')
```
